```python
s = "hello"
vowel="aeiou"
vowelc=0
consonant=0
for i in s:
    if i in vowel:
        vowelc=vowelc+1
    else:
        consonant=consonant+1
print("vowel{}consonant{}".format(vowelc,consonant))
```

```python
s="listen"
s1="silent"
if len(s) != len(s1):
    print("string is not anangram")
else:
    s1=sorted(s1);
    s=sorted(s);
    if(s==s1):
        print("word is anagram")
    else:
        print("word is not anagram")
    
```
