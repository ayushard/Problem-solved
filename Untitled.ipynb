```python
x = input("enter the string")
stack = []
for i in range(0,len(x)):
    if x[i]=="(" or x[i] =="{" or x== "[":
        stack.append(x[i])
    elif x[i]==")":
        if stack[-1]== "(":
            stack.pop()
        else:
            break
    elif x[i]=="}":
        if stack[-1]=="{":
            stack.pop()
        else:
            break
    elif x[i]=="]":
        if stack[-1]=="[":
            stack.pop()
        else:
            break
if len(stack)!=0:
    print("Not valid")
else:
    print("valid")
        
```

```python
arr=[5,3,8,6,2]
for i in range(1,len(arr)):
    key = arr[i]
    j=i-1
    while j>=0 and arr[j]>key:
        arr[j+1]=arr[j]
        j=j-1
    arr[j+1]=key
print(arr)
```
