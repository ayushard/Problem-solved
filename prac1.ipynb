```python
class Solution:
    def minMutation(self, startGene: str, endGene: str, bank):
        count = 0

        while len(bank) != 0:
            found = False

            for i in range(len(bank)):
                if sum(1 for c1, c2 in zip(startGene, bank[i]) if c1 != c2) == 1:
                    count += 1
                    startGene = bank[i]
                    bank.pop(i)

                    if startGene == endGene:
                        return count

                    found = True
                    break

            if not found:  # no valid next mutation
                return -1

        if startGene == endGene:  # EndGene -> endGene
            return count

        return -1
```

```python
class Solution:
    def myAtoi(self, s: str) -> int:
        s = s.strip()
        if not s: return 0 

        sign , i , num = 1 , 0 , 0 

        if s[0] in '+-':
            sign = -1 if s[0] == '-' else 1
            i = 1 

        while i < len(s) and s[i].isdigit():
            num = num * 10 + int(s[i])
            i += 1 
        return max(-2 ** 31 , min ( sign * num  , 2 ** 31 - 1))
```
