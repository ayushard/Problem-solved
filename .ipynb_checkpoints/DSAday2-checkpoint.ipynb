```python

name = "prashant"
c=""
for i in name:

    if i not in c:
        c=c+i;

print(c)
```

```python

name = "prashant"
c=""
for i in name[::-1]:

    if i not in c:
        c=c+i;

print(c)
```

```python

name = "prashant"
c=""
for i in range(7,0,-1):

    if name[i] not in c:
        c=c+name[i];

print(c)
```

```python
name = "prashant"
c=""
N=len(name)
for i in range(N-1,-1,-1):

    if name[i] not in c:
        c=c+name[i];

print(c)
```

```python
s = "hello"
vowel="aeiou"
vowelc=0
consonant=0
for i in s:
    if i in vowel:
        vowelc=vowelc+1
    else:
        consonant=consonant+1
print("vowel{}consonant{}".format(vowelc,consonant))
```

```python
s="listen"
s1="silent"
if len(s) != len(s1):
    print("string is not anangram")
else:
    s1=sorted(s1);
    s=sorted(s);
    if(s==s1):
        print("word is anagram")
    else:
        print("word is not anagram")
    
```

```python
s = "this is a sentence"
word=0
for i in range(0,len(s)-1):
    if s[i]==" ":
        if s[i-1]!=" " :
            word=word+1
print(word+1)
```
